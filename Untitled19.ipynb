```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(path):
    img = cv2.imread(path)

    if img is None:
        print("Error: image not loaded")
        return None, None

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img, (224, 224))
    img_denoised = cv2.GaussianBlur(img_resized, (3, 3), 0)
    img_norm = img_denoised / 255.0

    return img, img_norm
```

```python
original, processed = preprocess_image(img_path)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(original)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Processed")
plt.imshow(processed)
plt.axis('off')

plt.show()
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import os

input_folder = "/content/audio_raw"
output_folder = "/content/spectrogram_images"
os.makedirs(output_folder, exist_ok=True)

count = 0

for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".wav"):
            count += 1
            audio_file = os.path.join(root, file)

            audio, sr = librosa.load(audio_file, sr=16000)
            stft = np.abs(librosa.stft(audio))
            log_spec = librosa.amplitude_to_db(stft, ref=np.max)

            img_name = file.replace(".wav", "_spectrogram.png")
            img_path = os.path.join(output_folder, img_name)

            plt.figure(figsize=(5,5))
            librosa.display.specshow(log_spec, sr=sr, x_axis='time', y_axis='hz')
            plt.colorbar(format="%+2.0f dB")
            plt.title(file)
            plt.savefig(img_path)
            plt.close()

            print("✔ Saved:", img_path)

if count == 0:
    print("❌ No WAV files found!")
else:
    print("\n🎉 All spectrograms generated:", count)
```

```python
import os

wav_file = None
for root, dirs, files in os.walk("/content/audio_raw"):
    for f in files:
        if f.endswith(".wav"):
            wav_file = os.path.join(root, f)
            break

print("Selected file:", wav_file)
```

```python
import librosa
import noisereduce as nr

audio, sr = librosa.load(wav_file, sr=16000, mono=True)
audio_nr = nr.reduce_noise(y=audio, sr=sr)
audio_trimmed, _ = librosa.effects.trim(audio_nr, top_db=20)

print("Audio Loaded Successfully!")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

# Original waveform
plt.subplot(2,1,1)
plt.plot(audio)
plt.title("Original Waveform")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")

# Processed waveform
plt.subplot(2,1,2)
plt.plot(audio_trimmed)
plt.title("Processed Waveform (Noise Removed + Trimmed)")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
